# **1. Introduction**
---
- The **Premier League** is the **top tier of England's football pyramid**, with **20 teams** battling it out for the honour of being **crowned English champions**.

- Home to some of the most famous clubs, players, managers and stadiums in world football, the Premier League is the **most-watched league** on the planet with one billion homes watching the action in 188 countries.

- The league takes place between **August and May** and involves the teams playing each other **home and away** across the season, a total of **380 matches**.

- **Three points** are awarded for a **win**, **one point** for a **draw** and none for a defeat, with the team with the most points at the end of the season winning the Premier League title.

- The teams that finish in the **bottom three** of the league table at the end of the campaign are **relegated to the Championship**, the second tier of English football.

- Those teams are replaced by **three clubs promoted from the Championship**; the sides that finish in first and second place and the third via the end-of-season playoffs. 

- A total of **49 clubs** have played in the Premier League, with Brighton & Hove Albion and Huddersfield Town being the latest to make their debuts, in 2017/18.

- The competitiveness fo this league makes it the most watched league in the world.

- Since the League began in 1992, there have been seven different winners: Manchester United, Arsenal, Chelsea, Manchester City, Blackburn Rovers, Leicester City and Liverpool. Man Utd have had the most success with 13 titles in the 28 seasons so far. 

- **Current Champions -> Liverpool(2019-20 season).**

# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

- As we can see from our pre-profiling of data, the **Division column is a constant**. Hence we **drop it**.

```python
data.drop(['Div'], axis=1, inplace=True) # Insert your code here...
data.head()
```

- Checking for data entry mistake in **'HomeTeam' & 'AwayTeam'** Columns

```python
data['HomeTeam'].unique()
```

```python
data['AwayTeam'].unique()
```

- The unique() function shows us that the **'Middlesboro'** is a entry mistake, which should be **corrected to 'Middlesbrough'** in HomeTeam & AwayTeam.

```python
data['HomeTeam'] = data['HomeTeam'].str.replace('Middlesboro', 'Middlesbrough')
print(data['HomeTeam'].unique())
data['AwayTeam'] = data['AwayTeam'].str.replace('Middlesboro', 'Middlesbrough')
data['AwayTeam'].unique()
```

- **There are missing values in Halftime Data(HTHG, HTAG, HTR) which needs to be fixed.**

```python
hthg_median = data['HTHG'].median()
data['HTHG'].fillna(hthg_median, inplace=True)

htag_median = data['HTAG'].median()
data['HTAG'].fillna(htag_median, inplace=True)

data.head()
```

```python
data['HTR'].fillna('D', inplace=True)
data.head()
```

- **Now we will convert Date column to datetime Datatype for EDA.**

```python
from datetime import datetime
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
#data.iloc[152:160:,:]

data.info()
```

## **8.1 Seasonwise Breakdown**
---

```python
data.groupby('Season')['FTR'].value_counts().plot.bar(stacked=True, figsize=(21,10), color=['blue','red','yellow'])
plt.title('Seasonwise results analysis', fontsize=16);
plt.ylabel('No. of Games');
```

- There is a theory called 'Home Team Advantage', which can be seen in above graph. The number of Home wins is much higher in each season than Away wins or draws.

**Seasonwise Goals Scored by Home & Away Teams.**

```python
data.groupby('Season')['FTHG'].sum().plot.bar(figsize=(17,10), cmap='plasma')
data.groupby('Season')['FTAG'].sum().plot.bar(figsize=(17,10), cmap='autumn')
plt.title('Seasonwise Goals Scored Home & Away', fontsize=16);
plt.ylabel('No. of Goals Scored')
plt.legend(['HomeTeam Goals', 'AwayTeam Goals']);
```

```python
data.groupby('Season')['FTHG','FTAG'].sum()
```

- The above seasonwise Home & Away team goals again shows the Home team advantage.
- The 1993-94 & 1994-95 season had the highest number of goals scored home & away.
- Decrease in the goals scored after 1st 2 season is because of change of 22 teams to 20 teams per season.
- Total goals scored each season varies but the goals scored by home team is always more than Away goals. 

```python
data.groupby('Season')['HTR'].value_counts().plot.bar(stacked=True, figsize=(20,10), color=['yellow','red','blue'])
plt.title('Seasonwise Halftime results analysis', fontsize=16);
plt.ylabel('No. of games');
```

- Half time results clearly shows that the result at Halftime is a Draw mostly.
- 1st 2 season's halftime data was filled by Mode value of the column, hence it only shows Draw
